# __Sequential API allow yo to create models layer- by layer stacking them.__

__Import Libraries__

```python
import matplotlib.pyplot as plt
import tensorflow as tf 
from tensorflow.keras import layers
layer= tf.keras.layers 
import numpy as np 
```

__Load and Preprocess the Data__

```python
#Downloading mnist dataset(Rodrigue Nya) 

mnist = tf.keras.datasets.fashion_mnist 
```

```python
#Spliting data between train and test sets(Rodrigue Nya)

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_train, x_test =x_train/255.0, x_test/255.0
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

plt.figure(figsize= (10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_train[i], cmap= plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
    plt.show()
```

```python
#Building the model 

model = tf.keras.Sequential()
```

__Define the Model(By Rodrigue Nya)__

```python
model.add(layers.Flatten())
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(64, activation= 'relu'))
model.add(layers.Dense(10, activation= 'softmax'))
```

__Compile the Model and Training__

```python
#Model compilation 

model.compile(optimizer= 'adam', 
             loss= 'sparse_categorical_crossentropy', 
             metrics= ['accuracy'])
model.fit(x_train, y_train, epochs= 5)
```

__Model Evaluation( By Rodrigue Nya)__

```python
#Model Evaluation 
model.evaluate(x_test, y_test)
```

```python
#Model Prediction 
predictions = model.predict(x_test)
```

__Making Predictions__

```python
predicted_label= class_names[np.argmax(predictions[0])]
print('Actual label:', class_names[y_test[0]])
print('Predicted label:', predicted_label)
```

```python
predicted_label= class_names[np.argmax(predictions[2])]
print('Actual label:', class_names[y_test[2]])
print('Predicted label:', predicted_label)
```

__You can see that our model is making accurate prediction__.
